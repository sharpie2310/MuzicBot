```python
from bottle import route, run

@route('/irish')
def irish():
    f=open('/home/ubuntu/Project/Recipes/examples/Output/country_lstm_output_context_dropout_irish.txt').readlines()
    f=f[-150:]
    print(f)
    output=""
    started = False
    for line in f:
        if "<start>" in line:
            started=True
            continue
        if ("<end>" in line or "------" in line) and started:
            break
        if started:
            output+=line
    print(output)
    return output

@route('/swedish')
def swedish():
    f=open('/home/ubuntu/Project/Recipes/examples/Output/country_lstm_output_context_dropout_swedish.txt').readlines()
    f=f[-150:]
    print(f)
    output=""
    started = False
    for line in f:
        if "<start>" in line:
            started=True
            continue
        if ("<end>" in line or "------" in line) and started:
            break
        if started:
            output+=line
    print(output)
    return output

run(host='ec2-54-67-49-59.us-west-1.compute.amazonaws.com', port=8000, debug=True)
```
