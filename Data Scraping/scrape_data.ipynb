```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(urlopen("http://www.lotro-abc.com/songlist.html").read())
```

```python
prev = []
base_url = "http://www.lotro-abc.com/"
for row in soup('table'):
    tds = row('td')
    if (len(tds) > 1):
        for i in range(len(tds)):
            if('Rock/Pop' in tds[i]):
                if 'zip' in prev[0].find('a')['href']:
                    continue
                base_url += prev[0].find('a')['href']
                sp = BeautifulSoup(urlopen(base_url).read())
                base_url = "http://www.lotro-abc.com/"
                print(len(sp.find('p').getText()))
                if len(sp.find('p').getText()) < 800:
                    f = open('Rock_Pop.txt','a')
                    f.write(sp.find('p').getText())
                    f.write('\n')
                    f.close()
            prev = []
            prev.append(tds[i])
```
